### 연습 1. 딕셔너리 (Dictionary)

dictionary는 key와 value의 쌍으로 저장하는 자료구조입니다.

무조건 숫자 0부터 n-1까지 (또는 -1부터 -n까지) 인덱싱하는 리스트와는 달리 key로 접근합니다.

key는 숫자, 문자, 튜플 등 변경불가능한(immutable) 객체는 모두 가능해요.

그럼 예제를 통해 알아봅시다.

```python
# 빈 dictionary 생성의 두 가지 방법
dict_a = dict()
dict_b = {}

print(dict_a, type(dict_a))
print(dict_b, type(dict_b))


# 그럼 빈 list는 어떻게 했더라?
list_a = list()
list_b = []

print(list_a, type(list_a))
print(list_b, type(list_b))


# 내용이 있는 dictionary 생성  { key: value } 형태
dict_c = {
  'name': '와사비',
  'age': 30,
  '색깔': '연두색'
}

print(dict_c)
```

```python

# key로 접근해서 value 가져오기
name = dict_c['name']
age = dict_c['age']
print(name)
print(age)


# 새로운 key, value 추가 또는 기존 value 변경
dict_c['name'] = '애플망고'   # key가 이미 있으면 value가 변경됨
dict_c['taste'] = '달달'      # key가 없으면 새롭게 추가됨

print(dict_c)


# key는 튜플도 가능해요
dict_c[('색깔', '겉')] = '빨강'
dict_c[('색깔', '속')] = '노랑'

print(dict_c)


# 그런데 리스트는 안됩니다.
# dict_c[['색깔', '씨']] = '회색'   # <-- 이거 에러남

# 리스트는 내용이 변할 수 있기 때문이죠. (값수정, 추가, 삭제)
# 반면 튜플은 한번 정하면 내부 원소 값을 바꾸거나 원소를 추가하거나 제거할 수 없는 immutable type이예요.
```

### 연습 2. 딕셔너리 반복

리스트의 각 원소의 반복은  for item in some_list: ...

그렇다면 딕셔너리의 각 원소들을 반복할 때는 어떻게 할까요?

key를 불러오는 방법, value를 불러오는 방법, (key, value) 쌍으로 불러오는 방법, 이렇게 세 가지 방법이 있어요.

단, 딕셔너리는 순서가 따로 관리되지 않기 때문에 반복할 때 특정한 순서를 보장해주진 않아요.

```python
# key를 하나씩 불러오는 방법  (순서는 없어요. 딕셔너리 맘대로)
print('== key 반복 ==')

for key in dict_c.keys():
  print(key, ':', dict_c[key])

print()


# value를 하나씩 불러오는 방법
print('== value 반복 ==')

for value in dict_c.values():
  print(value)

print()


# (key, value) 쌍으로 한쌍씩 불러오는 방법 (튜플)
print('== (key, value) 반복 ==')

for key, value in dict_c.items():
  print(key, ':', value)

```

```python
# Q: 여기서 문제!!
# 리스트 처럼 그냥 
# 
# for something in some_dict:  
#   ...
# 
# 하면 어떻게 될까요?  keys(), values(), items() 생략하고요.
# 
# 한번 해보죠

for something in dict_c:
  print(something)


# 네, key가 나오네요. 즉,
# 
# for key in dict_c.keys():  
#   ...
# 
# 와 동일합니다. 생략하면 keys() 라는 것~! 기억해 두세요.
```

### 연습 3. 리스트에서 갯수 세기

0 ~ 1000 사이의 숫자 중에서 중복을 허용한 50개의 숫자를 담은 리스트가 있다고 가정해 볼께요.

이 리스트 안에 어떤 수가 몇 번씩 나오는지 세고 싶다고 할 때, 어떻게 하면 좋을까요?


1) "1000개짜리 카운터 리스트"를 만들고 0으로 채운 다음에, 50개짜리 숫자 리스트 내부를 반복하면서, 나타나는 원소를 "index"로 하여 카운터 리스트의 해당 값을 1씩 증가시킨다.

2) "비어있는 카운터 딕셔너리"를 만들고, 50개짜리 숫자 리스트 내부를 반복하면서, 나타나는 원소를 "key"로 하여 카운터 딕셔너리의 해당 값을 1씩 증가시킨다. (없으면 1로 추가하고)

어떤 것이 효율적일까요?

기껏해야 50개의 (중복되면 그것보다 적은) 숫자의 카운트를 저장하기 위해 1000개짜리 리스트를 만드는 것은 비효율이겠죠?

그리고 만약 세야 하는 것이 숫자가 아니라 사람 이름이라면??  범위를 알 수 없으니 카운트 리스트를 만드는 건 무리죠.

딕셔너리는 딱 필요한 만큼의 데이터 공간만 생성하므로 훨씬 효율적이라고 할 수 있습니다.

그럼 한번 볼까요?

```python
numbers = [15, 584, 699, 356, 784, 173, 609, 584, 152, 431, 890, 2, 513, 219, 988, 784, 584, 909, 528, 483, 369, 254, 289, 475, 512,
           784, 289, 35, 146, 302, 567, 120, 932, 214, 204, 948, 34, 215, 457, 569, 352, 290, 484, 369, 23, 2, 496, 831, 290, 385]


counter_dict = {}

for num in numbers:
  # key에 없으면 추가해줌
  if num not in counter_dict:    # if num not in counter_dict.keys() 와 동일!
    counter_dict[num] = 0
    
  # 해당 카운트를 1씩 증가시킴
  counter_dict[num] += 1

print(counter_dict)
print()


# 2개 이상인 것만 출력해보기
for num, count in counter_dict.items():
  if count >= 2:
    print(num, ':', count)
```

```python
# dictionary의 get() 함수와 두 번째 파라메터(default)를 이용하면 if문을 없앨 수도 있어요
counter_dict = {}

for num in numbers:
  counter_dict[num] = counter_dict.get(num, 0) + 1    # 찾아서 없으면 0을 리턴함, 그리고 1을 더함

print(counter_dict)
```

### 연습 5. 집합 (Set)

파이썬의 집합 자료형은 교집합, 합집합, 차집합 등의 집합 관련 연산을 수행하기에 편리한 자료형입니다.

중복을 허용하지 않기에 위에 언급한 집합 연산 수행이 간편하게 가능하죠.

별도로 순서를 관리하지는 않기 때문에 인덱싱은 안됩니다.


집합의 선언 방법은 set()으로 하거나 {1, 3, 5}와 같이 중괄호로 묶어서 선언할 수도 있는데,

주의할 점은 {} 이렇게 빈 중괄호를 쓰면 빈 집합이 아니라 빈 딕셔너리가 생깁니다. 

딕셔너리도 중괄호를 사용하니까 비어있으면 구분이 안되는데, 딕셔너리가 우선순위가 높다고 생각하면 됩니다.

```python
# 비어있는 집합의 생성
set_a = set()
print(set_a, type(set_a))


# 내용이 있는 집합은 중괄호 안에 나열하거나, 또는 set(...) 함수 안에 반복 가능한 객체를 넣자
set_b = {1, 3, 5, 7, 9}
set_c = set([1, 3, 5, 7, 9])

print(set_b, type(set_b))
print(set_c, type(set_c))


# 주의: 아래 두 개는 다르다. 왜 다른지 생각해 보자.
set_d = {'test'}
set_e = set('test')

print(set_d, type(set_d))
print(set_e, type(set_e))
```

```python
# 집합의 연산
set_f = {'John', 'Jane', 'James', 'Tom'}
set_g = {'Tom', 'Emma', 'Chris', 'Jennifer'}

set_inter = set_f & set_g
print(set_inter)

set_union = set_f | set_g
print(set_union)

set_diff = set_f - set_g
print(set_diff)


# 요소 추가
set_f.add('Emily')
print(set_f)

# 요소 삭제
set_f.remove('James')
print(set_f)
```

### 연습 6. CSV 파일 읽기

자, 이제 본격적으로 실습예제를 시작해 봅시다.

1회부터 999회까지의 로또 당첨번호가 담긴 csv 파일을 읽어들여 보겠습니다.

CSV는 Comma Separated Values 즉, 쉼표로 구분된 값들이라는 뜻입니다.

일단은 텍스트 파일이기 때문에, 먼저 줄 단위로 읽어서 문자열로 저장하고,

그 문자열을 쉼표 단위로 끊어서 여러 문자열로 나누는 작업을 해야 합니다.

그렇게 나뉜 조각 문자들을 숫자로 변환해야 사용하기 편하겠지요?

```python
# 텍스트 파일 읽는 방법 (내장함수 open() 사용)

file_obj = open('lotto999.csv', 'rt', encoding='utf-8')

# 줄 단위로 읽어서 배열로 가져오기. 배열의 각 원소는 각 줄에 해당
lines = file_obj.readlines()

# 다 쓴 후에는 닫아주어야 합니다. 이미 전체를 읽어서 lines 변수에 저장했으니 파일은 닫아도 됩니다.
file_obj.close()


# 첫째줄(index 0)을 출력해 봅시다. 컬럼의 헤더에 해당하므로 여기서는 이 첫째줄은 사용 안합니다.
print(lines[0])
```

### 연습 7. 문자열 분리(split), 숫자 변환

반복문을 통해 문자열을 한 줄씩 다루고, 쉼표를 기준으로 나누고, 이것을 숫자로 변환합니다.


```python

# 둘째줄(index 1)부터 for loop을 돌면서 각 줄의 텍스트를 쉼표(,)로 끊어 저장해 봅시다.
# 지금은 연습용이라서 다섯 라인 정도만 출력해 보겠습니다.
for line in lines[1:6]:
  line_strip = line.strip()           # 우측 끝에 줄바꿈문자(\n) 제거
  num_txts = line_strip.split(',')    # split() 함수를 통해 쉼표로 구분하여 배열로 저장
  print(num_txts)

print()


# 이번엔 숫자로 바꾸는 것까지 해보겠습니다.
for line in lines[1:6]:
  line_strip = line.strip()
  num_txts = line_strip.split(',')
  
  nums = []
  for txt in num_txts:
    nums.append(int(txt))
  
  print(nums)
```

### 연습 8. List Comprehension

파이썬의 독특한 문법 중 하나인 List Comprehension (또는 List Expression)을 연습해보고 가겠습니다.

인터프리터 언어인 파이썬에서는 성능향상을 위한 필수에 가깝거든요. 코드 양도 줄고요.

처음부터 줄어든 코드를 보면 어렵기에, 먼저 짜여진 반복문에서 출발하여 단계적으로 배워보도록 하겠습니다.

```python

# LC 1단계: 숫자로 바꾸는 세 줄을 List Comprehension으로 바꿔보겠습니다.
#        (연습용이므로 아직은 1~5번 라인만 가지고 합니다.)

for line in lines[1:6]:
  line_strip = line.strip()
  num_txts = line_strip.split(',')
  nums = [int(txt) for txt in num_txts]   # 이거
  print(nums)


# 결과는 같습니다. 단지 코드 줄 수가 줄었고, 파이썬이 조금 더 빠르게 실행하는 것 뿐입니다. 
# 물론 5줄이나 1000줄 정도에서는 눈에 띄는 속도 차이는 없습니다.
```

```python

# LC 2단계: 1회성인 변수 선언을 없애고, 더 줄여보겠습니다.
for line in lines[1:6]:
  nums = [int(txt) for txt in line.strip().split(',')]
  print(nums)

```

```python
# 여기서 잠깐!
# 현재는 화면에 출력만 되는데, 무언가 자료구조에 담아줘야 나중에 사용 가능하겠죠?
# 상위의 리스트를 하나 만들어 보겠습니다.

history_data_list = []
for line in lines[1:6]:
  nums = [int(txt) for txt in line.strip().split(',')]
  history_data_list.append(nums)
  
print(*history_data_list, sep='\n')   # 한 줄씩 출력하기 위해서 기교를 좀 부려봤습니다. 잘 모르겠으면 지금은 그냥 넘어가셔도 됩니다.
```

```python
# LC 3단계: 이제 최종적으로 줄여볼까요?
history_data_list = [[int(txt) for txt in line.strip().split(',')] for line in lines[1:6]]


print(*history_data_list, sep='\n')


# 이렇게 2중 for 문도 List Comprehension으로 표현할 수 있습니다.
# 지금 완벽히 이해가 안돼도 괜찮습니다.
# 다만 이렇게 단계적으로 줄여보고, 많이 써보다 보면, 언젠간 익숙해지는 순간이 옵니다.
```

```python
# 자 이제 5 라인이 아니라 전체 999 라인을 담아 보도록 하겠습니다. (헤더 한 라인만 제외하고요)
# lines[1:6]을 limit 해제하여 lines[1:] 으로 바꾸면 됩니다.

history_data_list = [[int(txt) for txt in line.strip().split(',')] for line in lines[1:]]

print(*history_data_list, sep='\n')
```
